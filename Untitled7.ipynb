````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

class FakeNewsDetectorPipeline:
    def __init__(self, stop_words='english', max_df=0.7, max_iter=50):
        self.vectorizer = TfidfVectorizer(stop_words=stop_words, max_df=max_df)
        self.model = PassiveAggressiveClassifier(max_iter=max_iter, random_state=42)
        self.is_trained = False

    def prepare_data(self, df, text_col, label_col, test_size=0.2):
        X = df[text_col].fillna('')
        y = df[label_col]
        return train_test_split(X, y, test_size=test_size, random_state=42)

    def fit(self, X_train, y_train):
        X_train_vectorized = self.vectorizer.fit_transform(X_train)
        self.model.fit(X_train_vectorized, y_train)
        self.is_trained = True
        return self

    def predict(self, X_test):
        if not self.is_trained:
            raise ValueError("Pipeline configuration error: Model is not trained yet.")
        X_test_vectorized = self.vectorizer.transform(X_test)
        return self.model.predict(X_test_vectorized)

    def evaluate(self, y_true, y_pred):
        performance = {
            "accuracy": accuracy_score(y_true, y_pred),
            "confusion_matrix": confusion_matrix(y_true, y_pred),
            "classification_report": classification_report(y_true, y_pred, output_dict=True)
        }
        return performance

    def infer(self, custom_text):
        if not self.is_trained:
            raise ValueError("Pipeline execution failure: Model has no active weights.")
        vectorized_text = self.vectorizer.transform([custom_text])
        return self.model.predict(vectorized_text)[0]

try:
    pipeline = FakeNewsDetectorPipeline()
    X_train, X_test, y_train, y_test = pipeline.prepare_data(df, text_col='text', label_col='label')
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
    metrics = pipeline.evaluate(y_test, predictions)

    print(f"=========================================")
    print(f" PIPELINE PERFORMANCE VERIFICATION SUCCESS")
    print(f"=========================================")
    print(f"Overall Accuracy Metrics: {metrics['accuracy'] * 100:.2f}%\n")
    print("Classification Matrix Breakdown:")
    print(pd.DataFrame(metrics['confusion_matrix'], index=['Actual FAKE', 'Actual REAL'], columns=['Predicted FAKE', 'Predicted REAL']))
except Exception as e:
    print(f"Pipeline Execution Aborted: {str(e)}")
````

````python
import io
import numpy as np
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

class FakeNewsDetectorPipeline:
    def __init__(self, stop_words='english', max_df=0.7, max_iter=50):
        self.vectorizer = TfidfVectorizer(stop_words=stop_words, max_df=max_df)
        self.model = PassiveAggressiveClassifier(max_iter=max_iter, random_state=42)
        self.is_trained = False

    def prepare_data(self, df, text_col, label_col, test_size=0.2):
        X = df[text_col].fillna('')
        y = df[label_col]
        return train_test_split(X, y, test_size=test_size, random_state=42)

    def fit(self, X_train, y_train):
        X_train_vectorized = self.vectorizer.fit_transform(X_train)
        self.model.fit(X_train_vectorized, y_train)
        self.is_trained = True
        return self

    def predict(self, X_test):
        if not self.is_trained:
            raise ValueError("Pipeline configuration error: Model is not trained yet.")
        X_test_vectorized = self.vectorizer.transform(X_test)
        return self.model.predict(X_test_vectorized)

    def evaluate(self, y_true, y_pred):
        return {
            "accuracy": accuracy_score(y_true, y_pred),
            "confusion_matrix": confusion_matrix(y_true, y_pred)
        }

    def infer(self, custom_text):
        if not self.is_trained:
            raise ValueError("Pipeline execution failure: Model has no active weights.")
        vectorized_text = self.vectorizer.transform([custom_text])
        return self.model.predict(vectorized_text)[0]

try:
    if 'df' not in locals():
        uploaded = files.upload()
        filename = list(uploaded.keys())[0]
        df = pd.read_csv(io.BytesIO(uploaded[filename]))

    pipeline = FakeNewsDetectorPipeline()
    X_train, X_test, y_train, y_test = pipeline.prepare_data(df, text_col='text', label_col='label')
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
    metrics = pipeline.evaluate(y_test, predictions)

    print(f"Accuracy: {metrics['accuracy'] * 100:.2f}%")
    print("Confusion Matrix:")
    print(metrics['confusion_matrix'])

    test_news = "BREAKING: NASA just confirmed that a massive asteroid will hit New York tomorrow morning."
    prediction = pipeline.infer(test_news)
    print(f"Prediction for custom text: {prediction}")

except Exception as e:
    print(f"Error: {str(e)}")
````

````python
with open('app.py', 'w') as f:
    f.write("""
import streamlit as st
import pickle
import re
import time
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

with open('model.pkl', 'rb') as f_model:
    model = pickle.load(f_model)

with open('vectorizer.pkl', 'rb') as f_vec:
    vectorizer = pickle.load(f_vec)

def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text.strip().lower()

st.set_page_config(page_title="Fake News Detector Pro", layout="centered")
st.title("Fake News Detection Portal")
st.subheader("Enterprise-grade Machine Learning Verification Engine")

user_input = st.text_area("Paste News Content Here", height=200)

if st.button("Verify News"):
    if user_input.strip() != "":
        start_time = time.time()

        cleaned_input = clean_text(user_input)
        vectorized_text = vectorizer.transform([cleaned_input])
        prediction = model.predict(vectorized_text)[0]

        execution_time = (time.time() - start_time) * 1000

        logging.info(f"Prediction: {prediction} | Latency: {execution_time:.2f}ms")

        if prediction == "REAL":
            st.success(f"Result: The News is REAL (Processed in {execution_time:.2f}ms)")
        else:
            st.error(f"Result: The News is FAKE (Processed in {execution_time:.2f}ms)")
    else:
        st.warning("Please enter some text to verify.")
""")
````
